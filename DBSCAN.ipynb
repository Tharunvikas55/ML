```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import DBSCAN
from sklearn .preprocessing import StandardScaler
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df =pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
df
```

```python
features =['Annual Income (k$)','Spending Score (1-100)']
x =df[features]
```

```python
x_standardized =StandardScaler().fit_transform(x)
```

```python
dbscan =DBSCAN(eps=0.5,min_samples =5)
clusters =dbscan.fit_predict(x_standardized)
```

```python
centroids =pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'],c=clusters,cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending score')
plt.show()
```
